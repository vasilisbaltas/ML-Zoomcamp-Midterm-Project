```python
import pandas as pd
pd.options.display.max_rows = 200
import numpy as np
import matplotlib.pyplot as plt
```

The following dataset contains the historical sales of cars in Norway from January 2007 to January 2017 and you can download it
 on supchains.com/download  --> NORWAY CAR SALES

```python
cars = pd.read_csv('norway_new_car_sales_by_make1.csv')
cars.head()
```

```python
cars.info()                         # our dataset containes only 10 missing values in the Make column
cars.isna().sum()
```

So our dataset contains car historical sales from 2007 to 2017

```python
cars['Make'].nunique()
```

```python
# our dataset includes 65 car brands
cars['Make'].unique()
```

```python
# we can also create a new datetime column in order to plot the time series and visually inspect them
cars['Date'] = pd.to_datetime(cars['Year'].astype(str) + '-' + cars['Month'].astype(str), format='%Y-%m')
```

```python
# let's make a dataframe that contains only the Audi sales
audi = cars[cars['Make'] == 'Audi']
```

```python
# let's check the behaviour of this time series
plt.figure(figsize=(10, 6))
plt.plot(audi['Date'], audi['Quantity'], color='navy')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.title('Audi Sales')
plt.show()
```

Of course we can plot the sales of any car brand that is of interest to us
